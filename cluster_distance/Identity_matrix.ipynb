```python
%%time
import pandas as pd
from sklearn import metrics,cluster
import re
import numpy as np

#### Reading the feature and calculating the pairwise distance between all the points

df = pd.read_csv('gh_lyso_PAAC.txt', sep='\t')# input file which contains features, header=None if no column title
x=df.iloc[:,1:].values
distance=metrics.pairwise.pairwise_distances(x)
y=df.iloc[:,0].values
data,labe=[],[]
with open('pairwise_identity_gh_lyso.txt','r') as inpt:
    for each in inpt:
        r=re.split('\s+',each.strip())
        data.append(r[2:])
        labe.append(r[1])
df1=pd.DataFrame(data,index=labe,columns=labe)
df1.to_excel('Seq_pairwise_distance_identity_matrix.xlsx')
df1.to_csv('Seq_pairwise_distance_identity_matrix.csv')
```

```python
tot=len(distance)
sub_tot=len(distance[0])
identity=np.zeros([tot,sub_tot])
for g in range(tot):
    for h in range(sub_tot):
        identity[g][h]=round((1-distance[g][h])*100,2)
df2=pd.DataFrame(identity,index=y,columns=y)
df2.to_excel('PAAC_pairwise_distance_identity_matrix.xlsx')
```
